```python
import numpy as np

from scipy.stats import norm
from scipy.stats import lognorm
from scipy.stats import bernoulli
from scipy.stats import expon
import seaborn as sns
```

```python
# Num workers
def worker():
    vars = np.round(lognorm.rvs(.6, loc=5 , scale=10, size=1000))
    vars = np.where(vars > 35, 35, vars)
    sns.displot(vars, binwidth=1)
worker()
```

```python
def days():
    vars = np.round(lognorm.rvs(1, loc=1 , scale=3, size=1000))
    vars = np.where(vars > 10, 10, vars)
    sns.displot(vars, binwidth=1)
days()
```

```python
def shifts():
    #vars = np.round(lognorm.rvs(2, loc=4, scale=1.5, size=1000))
    #vars = np.where(vars > 12, 12, vars)
    vars = np.round(norm.rvs(loc=7, scale=1.5, size=1000))
    vars = np.where(vars > 12, 12, vars)
    vars = np.where(vars < 4, 4, vars)
    sns.displot(vars, binwidth=1)
shifts()
```

```python
def tracks():
    vars = np.round(1 + lognorm.rvs(1, scale=1, size=1000))
    vars = np.where(vars > 6, 6, vars)
    sns.displot(vars, binwidth=1)

tracks()
```

```python
def availability():
    #vars = (lognorm.rvs(1, loc=0, scale=.1, size=1000))
    vars = 1 - expon.rvs(scale=.1, size=1000)
    vars = np.where(vars < 0.7, 0.7, vars)
    sns.displot(vars, binwidth=.05)

availability()
```

```python
def avg_req():
    vars = norm.rvs(loc=0.5, scale=0.15, size=10000)
    vars = np.where(vars > 1, 1, vars)
    vars = np.where(vars <= 0, 0, vars)
    sns.displot(vars, binwidth=0.01)
avg_req()
```
