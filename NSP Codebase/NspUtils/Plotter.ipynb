```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/thomas/Dropbox/FYP/FypCodebase/results.csv')
df.head()
```

```python
df['Exact Solve Time'] = df['Exact Solve Time']/60
df.head()
```

```python
plot = sns.scatterplot(data=df, x='pred_prune_pc', y='solve_time_diff_pc',
                       size='Exact Solve Time', alpha=0.5, sizes=(5, 800))
plot.set(xlabel='Prune Percentage', ylabel='Running Time Reduction',
         title=f'')
```

```python
fig = plot.get_figure()
fig.savefig(f'results.png', dpi=300)
```
